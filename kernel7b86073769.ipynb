```python
# %% [code]
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score
# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# Any results you write to the current directory are saved as output.

# %% [code]
deliveries=pd.read_csv("../input/ipl/deliveries.csv")
matches=pd.read_csv("../input/ipl/matches.csv")

myimputer=SimpleImputer(strategy="most_frequent")
cols_with_missing_value=[]
matches.drop(["id","date","venue","umpire3",'date'],axis=1,inplace=True)
cols_with_missing_values=[]
for col in matches.columns:
    if matches[col].isnull().any():
        cols_with_missing_values.append(col)

   

    
for col in cols_with_missing_value:
    matches[col+"_ismissing"]=matches[col].isnull()
    #matches.drop(col,axis=1)
match_columns=matches.columns
matches=pd.DataFrame(myimputer.fit_transform(matches))       
matches.columns=match_columns
for col in matches.columns:
    if matches[col].dtype=="object":
        #print(col)
        #print(matches[col].dtype)
        #print(matches[col].unique())
        #matchescategorical=pd.DataFrame.append(matches[col].astype('category'))
        matches[col]=matches[col].astype('category',inplace=True)
        #print(col)
        matches[col+"_cat"]=matches[col].cat.codes
        matches.drop([col],axis=1,inplace=True)

        
#how to know data was categorical




#print(cols_with_missing_values)

X=matches.drop("winner_cat",axis=1)
y=matches["winner_cat"]



x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)

#x_train = x_train[:, None]
forest_model=RandomForestRegressor(n_estimators=50,max_features=4,random_state=1)
#print(str(x_test.shape)+"   "+str(y_test.shape))
forest_model.fit(x_train,y_train)
preds=forest_model.predict(x_test)
print(mean_absolute_error(y_test,preds))
print(accuracy_score(preds,y_test,normalize=False))






# %% [code]
```
